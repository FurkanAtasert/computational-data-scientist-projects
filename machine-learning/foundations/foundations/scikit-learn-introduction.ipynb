## Using the Scikit-learn Dataset

Scikit-learn comes with several standard sample datasets that make learning machine learning easier. To load these sample datasets, import the datasets module and load the dataset you want. For example, the following code snippets load the Iris dataset:

```python
from sklearn import datasets
iris = datasets.load_iris()
```

Tip: The Iris flower dataset, or Fisher’s Iris dataset, is a multivariate dataset introduced by the British statistician and biologist Ronald Fisher. The dataset consists of 50 samples from each of three Iris species (Iris setosa, Iris virginica, and Iris versicolor). For each sample, the length and width of the sepals and petals were measured in centimeters. Based on this combination of four features, Fisher developed a linear discriminant model to distinguish the species from one another.

The loaded dataset is represented as a Bunch object, which is a Python dictionary that provides attribute-style access. You can use the DESCR attribute to obtain a description of the dataset.

```python
print(iris.DESCR)
```

However, you can obtain the features of the dataset by using the data attribute:

```python
print(iris.data)
```

You can also use the feature_names attribute to print the names of the features.

```python
print(iris.feature_names)
```

The statement above prints the following:

This means that the dataset contains four columns—the sepal length, sepal width, petal length, and petal width. If you are wondering what petals and sepals are, Figure 5.1 shows the four-part flower of Ludwigia octovalvis, illustrating the petals and sepals (Source: https://en.wikipedia.org/wiki/Sepal
).

To print the labels of the dataset, use the target attribute. For the label names, use the target_names attribute:

```python
print("Labels: \n",iris.target,"\n")
print("Label Names: ",iris.target_names)
```

This prints the following:

In this case, 0 represents setosa, 1 represents versicolor, and 2 represents virginica.

Tip: Note that not all sample datasets in Scikit-learn support the feature_names and target_names attributes.

Figure 5.2 summarizes what the dataset looks like.

```python
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
import numpy as np

plt.figure(figsize=(8, 5))
plt.scatter(
    iris.data[:, 0],
    iris.data[:, 1],
    c=iris.target
)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Iris Dataset")
plt.show()
```

## Creating Your Own Dataset

If you cannot find a suitable dataset for experimentation, why not create your own dataset?

The sklearn.datasets.samples_generator module of the Scikit-learn library contains a set of functions that allow you to generate different types of datasets for different types of problems.

You can use it to create datasets such as:

- Linearly distributed datasets

- Clustered datasets

- Circularly clustered datasets

### Linearly Distributed Dataset

The make_regression() function generates linearly distributed data. You can specify the number of features you want and the standard deviation of the Gaussian noise applied to the output:

```python
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
import numpy as np

x, y = make_regression(n_samples=100, n_features=1, noise=5.4)
plt.scatter(x,y)
plt.show()
```

Figure 5.3 shows the distribution plot of the generated dataset.

### Clustered Dataset

The make_blobs() function generates n clusters of a random dataset. This is very useful when performing clustering in unsupervised learning (Chapter 9, “Supervised Learning—Classification Using K-Nearest Neighbors (KNN)”).

```python
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
import numpy as np

x, y = make_regression(n_samples=100, n_features=1, noise=5.4)
plt.scatter(x,y)
plt.show()
```

**Figure 5.3:** Distribution plot of the generated dataset.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
X, y = make_blobs(500, centers=3)
rgb = np.array(['r', 'g', 'b'])
plt.scatter(X[:, 0], X[:, 1], color=rgb[y])
```

### Circularly Clustered Dataset

The make_circles() function generates a random dataset in which a smaller circle is placed inside a larger circle in two dimensions. This is useful when performing classification using algorithms such as SVM (Support Vector Machines). SVM will be discussed in Chapter 8, “Supervised Learning — Classification Using SVM.”

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_circles
X, y = make_circles(n_samples=100, noise=0.09)
rgb = np.array(['r', 'g', 'b'])
plt.scatter(X[:, 0], X[:, 1], color=rgb[y])
```

**Figure 5.5** shows the distribution plot of the generated random dataset.
