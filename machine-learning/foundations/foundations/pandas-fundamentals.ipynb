# Manipulating Tabular Data Using Pandas

## What Is Pandas?

Although NumPy arrays are a greatly improved N-dimensional array object compared to Python lists, they are still insufficient to meet the needs of data science. In the real world, data is often presented in tabular format. For example, consider the contents of the following CSV file:

A CSV file contains rows of data divided into three columns — the index, the date and time of the record, and blood glucose measurements in mmol/L. To work effectively with data stored in tabular form, you need a more suitable data type — Pandas. While Python supports lists and dictionaries for handling structured data, they are not well suited for processing numerical tables such as those stored in a CSV file. Pandas is a Python package that provides fast, flexible, and expressive data structures designed to make working with “relational” or “labeled” data both easy and intuitive.

NOTE: Pandas stands for Panel Data Analysis.

Pandas supports two main data structures: Series and DataFrame. In this section, you will learn how to work with Series and DataFrames in Pandas.

## Pandas Series

A Pandas Series is a one-dimensional NumPy-like array in which each element has an index (by default 0, 1, 2, ...); a Series behaves very much like a dictionary containing an index. Figure 3.1 shows the structure of a Series in Pandas.

To create a Series, you first need to import the pandas library (by convention, the alias pd is used), and then use the Series class:

```python
import pandas as pd
seri = pd.Series([1,2,3,4,5])
print(seri)
```

The code snippet above will print the following:

```python
print(seri)
```

By default, the index of a Series starts at 0.

## Creating a Series with a Specific Index

You can use the Series function by specifying an optional index for a Series:

```python
seri = pd.Series([1,2,3,4,5], index=['a','b','c','d','c'])
print(seri)
```

The code snippet above will print the following:

```python
print(seri)
```

As seen in the previous output, the index of a Series does not have to be unique.

## Accessing Elements in a Series

Accessing an element in a Series is similar to accessing an element in an array. You can use the position of the element as follows:

```python
print(seri[2])
print(seri.iloc[2])
```

The iloc indexer allows you to specify an element by its position.

Alternatively, you can also specify the index value of the element you want to access, as follows:

```python
print(seri['d'])
print(seri.loc['d'])
```

The loc indexer allows you to specify the label (value) of an index. In both of the previous examples, the result is an integer (the type of this Series). What happens if we do the following?

```python
print(seri['c'])
```

In this case, the result will be another Series:

```python
print(seri['c'])
```

You can also perform slicing on a Series:

```python
print(seri[2:],"\n")
print(seri.iloc[2:])
```

The previous code snippet produces the following output:

```python
print(seri[2:],"\n")
print(seri.iloc[2:])
```

## Date Ranges

In the previous section, you saw how to create date ranges using the date_range() function. The periods parameter specifies the number of dates you want to generate, and the default frequency is set to D (Daily). If you want to change the frequency to monthly, set the freq parameter to M:

```python
tarihler2 = pd.date_range('2019-05-01', periods=12, freq='M')
print(tarihler2)
```

This will print the following dates:

When the frequency is set to monthly, the day of each date will be the last day of the month. If you want the date to start on the first day of the month, set the freq parameter to MS:

```python
tarihler2 = pd.date_range('2019-05-01', periods=12, freq='MS')
print(tarihler2)
```

Now you will see that each date starts on the first day of each month:

Tip: For other date frequencies, refer to the Offset Aliases section of the documentation:
http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases

Pandas automatically interprets the date you specify. In this case, 2019-05-01 is interpreted as May 1, 2019. In some regions, developers specify dates in the dd-mm-yyyy format. Therefore, to represent January 5, 2019, you should specify it as follows:

```python
tarihler2 = pd.date_range('05-01-2019', periods=12, freq='MS')
print(tarihler2)
```

However, in this case, Pandas will interpret 05 as the month, 01 as the day, and 2019 as the year, as shown in the following output:

In addition to setting the date, you can also specify the time:

```python
tarihler3 = pd.date_range('2019/05/17 09:00:00', periods=8, freq='H')
print(tarihler3)
```

You should see the following output:

If you are unsure, printing the date range is always useful for verification.

A DataFrame is highly useful in the world of data science and machine learning because it has a structure that closely resembles how data is stored in real life. Imagine data stored in a spreadsheet, and you will get a very good visual impression of what a DataFrame looks like. A Pandas DataFrame is frequently used to represent data in machine learning. In the remaining parts of this section, we will spend considerable time and effort understanding how a DataFrame works.

## Extracting Data from a DataFrame

In Chapter 2, “Extending Python with NumPy,” you learned about NumPy and how to extract part of a NumPy array using slicing. Similarly, in Pandas, slicing applies to both Series and DataFrames.

## Checking Whether a Result Is a DataFrame or a Series

One common issue you may encounter with Pandas is not knowing whether the result you obtained is a Series or a DataFrame. To resolve this uncertainty, the following function can help you:

```python
import pandas as pd

def checkSeriesOrDataframe(var): 
    if isinstance(var, pd.DataFrame): 
        return 'DataFrame' 
    if isinstance(var, pd.Series): 
        return 'Series'
```

## Sorting Data in a DataFrame

You can sort the data in a DataFrame in two ways:

- Sorting by labels (axis) using the sort_index() function.
- Sorting by values using the sort_values() function.

### Sorting by Index

To sort using an axis, you must specify whether you want to sort by index or by column. If you set the axis parameter to 0, it indicates that you want to sort by the index.

```python
print(df.sort_index(axis=0, ascending=False)) 
```

According to the statement above, the DataFrame is now sorted in descending order by index.

Note: The sort_index() function returns the sorted DataFrame. The original DataFrame is not affected. If you want the original DataFrame to be sorted, set the inplace parameter to True. In general, operations performed on a DataFrame usually do not modify the original DataFrame. Therefore, inplace is set to False by default. When inplace is set to True, the function returns None as the result.

If you set the axis parameter to 1, it indicates that you want to sort by the column labels.

```python
print(df.sort_index(axis=1, ascending=False))
```

The DataFrame is now sorted by column labels in descending order.

### Sorting by Values

To sort by values, use the sort_values() function. The following statement sorts the DataFrame by the values in column "A":

```python
print(df.sort_values('A', axis=0))
```

The output is now sorted in ascending order according to the values in column "A".

## Adding and Removing Rows and Columns in a DataFrame

So far, the previous sections focused on extracting rows and columns from DataFrames and how to sort DataFrames. In this section, we will focus on how to add and remove columns in DataFrames.

The following code snippet creates a DataFrame from a dictionary:

```python
import pandas as pd 
data = {'name': ['Janet', 'Nad', 'Timothy', 'June', 'Amy'], 
        'year': [2012, 2012, 2013, 2014, 2014], 
        'reports': [6, 13, 14, 1, 7]} 
df = pd.DataFrame(data, index = 
                  ['Singapore', 'China', 'Japan', 'Sweden', 'Norway']) 
print(df) 
```

The DataFrame is as follows:

### Adding a Column

The following code snippet demonstrates how to add a new column named "school" to the DataFrame:

```python
import numpy as np 
schools = np.array(["Cambridge","Oxford","Oxford","Cambridge","Oxford"]) 
df["school"] = schools 
print(df) 
```

Printing the DataFrame will look like this:

### Removing Rows

To remove one or more rows, use the drop() function. The following code snippet removes two rows with the index values "China" and "Japan":

```python
print(df.drop(['China', 'Japan']))
```

The following output shows that the two rows have been removed.

Tip: Like the sort_index() function, the drop() function does not affect the original DataFrame by default. If you want to modify the original DataFrame, use the inplace parameter.

If you want to remove a row based on a specific column value, specify the column name and the condition as follows:

```python
print(df[df.name != 'Nad'])
```

The statement above removes the row with the value "Nad".

You can also remove a row by its row number.

```python
print(df.drop(df.index[1])) 
```

The statement above removes row number 1 (the second row).

Since df.index[1] returns the value "China", the statement above is equivalent to df.drop['China'].

If you want to remove multiple rows, specify the row numbers as a list:

```python
print(df.drop(df.index[[1,2]])) 
```

The statement above removes rows 1 and 2 (the second and third rows).

The following removes the rows from the second position to the last position.

```python
print(df.drop(df.index[-2]))
```

You should see the following output:

### Removing Columns

The drop() function removes rows by default, but if you want to remove columns, set the axis parameter to 1.

```python
print(df.drop('reports', axis=1)) # drop column
```

The code snippet above removes the "reports" column.

You can also remove a column by its column number.

```python
print(df.drop(df.columns[1], axis=1))
```

This will remove the second column ("reports").

To remove multiple columns:

```python
print(df.drop(df.columns[[1,3]], axis=1))
```

This will remove the second and fourth columns ("reports" and "school").

## Creating a Cross Tabulation

In statistics, a cross tabulation is used to summarize the distribution of two or more variables and to show the relationships between these variables. Consider the following example:

```python
import pandas as pd

df = pd.DataFrame( 
 { 
 "Gender": ['Male','Male','Female','Female','Female'], 
 "Team" : [1,2,3,3,1] 
 }) 

print(df)
```

When you create a DataFrame using a dictionary and print the DataFrame, you will see an output like the following:

This DataFrame shows each person's gender and which team the person belongs to. By using a cross tabulation, you can summarize the distribution of each gender within each team and create a table. For this, you use the crosstab() function:

```python
print("Showing the gender distribution in each team")
print(pd.crosstab(df.Gender, df.Team))
```

```python
print(pd.crosstab(df.Team, df.Gender))
```
