Durum uzayı $\Omega$ ile gösterilsin. $A \subset \Omega$ ise $A$ kümesi üzerinde bir olasılık ölçüsü, $P:A \rightarrow [0,1]$ aralığına giden bir fonksiyondur ve aşağıdaki aksiyomları sağlar.
- $P(\emptyset) = 0$ ve $P(\Omega) = 1$
- $\forall_i \neq j$ için $A_i \cap A_j$ $\Rightarrow P(U^{\infty}_{i = 1}A_i) = \Sigma^{\infty}_{i = 1} P(A_i)$
- $P(A) = 1 - P(A)$
- $A \subseteq B$ $\Rightarrow$ $P(A) \leq P(B)$
- $P(A \cup B) = P(A) + P(B) - P(A \cap B), P(A \cap B) = P(A)P(B) \Rightarrow A \perp B$
- B olayı verilirse A olayının gerçekleşme olasılığı: $P(A/B) = \frac{P(A \cap B}{P(B)}$
- $P(A/B)P(B) = P(A \cap B) = P(B/A)P(A)$

Naive Bayes, sınıflandırıcılar oluşturmak için basit bir tekniktir. Bu tür sınıflandırıcıları eğitmek için tek bir algoritma yoktur, ancak ortak bir ilkeye dayanan bir algoritma ailesi vardır: tüm naive Bayes sınıflandırıcıları, belirli bir özelliğin değerinin, sınıf değişkeni göz önüne alındığında, başka herhangi bir özelliğin değerinden bağımsız olduğunu varsayar. Örneğin, bir meyve kırmızı, yuvarlak ve yaklaşık 10 cm çapında bir elma olarak kabul edilebilir. Naif bir Bayes sınıflandırıcısı, bu özelliklerin her birinin, renk, yuvarlaklık ve çap özellikleri arasındaki olası korelasyonlardan bağımsız olarak, bu meyvenin bir elma olma olasılığına bağımsız olarak katkıda bulunduğunu düşünmektedir.

Bazı olasılık modelleri türleri için, naive Bayes sınıflandırıcıları denetimli bir öğrenme ortamında çok verimli bir şekilde eğitilebilir. Birçok pratik uygulamada, naive Bayes modelleri için parametre tahmini, maksimum olasılık yöntemini kullanır; başka bir deyişle, Bayesci olasılığını kabul etmeden veya herhangi bir Bayesci yöntemi kullanmadan naive Bayes modeli ile çalışabilir. Basit tasarımlarına ve görünüşte aşırı basitleştirilmiş varsayımlarına rağmen, naive Bayes sınıflandırıcıları birçok karmaşık gerçek dünya durumunda oldukça iyi çalışmaktadır. Naive Bayes'in bir avantajı, sınıflandırma için gerekli parametreleri tahmin etmek için daha az sayıda eğitim verisine ihtiyaç duymasıdır.

### Naive Bayes algoritmalarının uygulamaları

**Gerçek zamanlı tahmin:** Naive Bayes, kullanışlı bir öğrenme sınıflandırıcısıdır ve kesinlikle hızlıdır. Böylece, gerçek zamanlı olarak tahminler yapmak için kullanılabilir.

**Çok sınıflı tahmin:** Bu algoritma aynı zamanda çok sınıflı tahmin özelliği ile de bilinir. Burada, birden fazla hedef değişken sınıfının olasılığını tahmin edebilir.

**Metin Sınıflandırması/ Spam filtreleme/ duyarlılık analizi:** Metin sınıflandırmasında kullanılan Naive Bayes sınıflandırıcıları, bağımsızlık altında çok sınıflı problemlerde daha iyi sonuç vermesi nedeniyle diğer algoritmalara kıyasla daha yüksek başarı oranına sahiptir. Spam filtrelemede (spam e-postasını tanımlama) ve duygu analizinde (sosyal medya analizinde, olumlu ve olumsuz müşteri duygularını tanımlamak için) yaygın olarak kullanılmaktadır.

**Öneri sistemi:** Naive Bayes Sınıflandırıcısı ve filtreleme, görünmeyen bilgileri filtrelemek ve bir kullanıcının belirli bir kaynağı isteyip istemediğini tahmin etmek için makine öğrenimi ve veri madenciliği tekniklerini kullanan bir öneri sistemi oluşturur.

### Naive Bayes Türleri 

**Gaussian Naive Bayes:** Eğer özelliklerimiz sürekli değer (continuous value) ise bu değerlerin
bir gauss dağılımı veya diğer bir değişle normal dağılımdan örneklendiğini varsayıyoruz.
Değerlerin veri kümesinde bulunma şekli değiştiğinden, koşullu olasılık formülü şu şekilde
değişir:

$P(x_i | y) = \frac{1}{\sqrt{2 \pi \sigma^2_y}} exp (-\frac{(x_i - \mu_y)^2}{2 \sigma^2_y})$

**Multinominal Naive Bayes:** Örneğin bir e-ticaret siteniz varsa pek çok kategoriniz olur. Çok
sınıflı kategorileri -spor, kozmetik, giyim, petshop- sınıflandırmak için kullanılır.
**Bernoulli Naive Bayes:** Multinominal Naive Bayes‟e benzer şekilde sınıflandırma yapar.
Ancak tahminler sadece boolean (ikili) şekildedir. Evet/Hayır, Spam/Spam Değil, 1/0,
Var/Yok gibi.

### OLASILIK MODELİ:

Bir sınıflama problemi olduğunda Naive Bayes bir koşullu olasılık modelidir. 
$x = (x_1, ... , x_n$ n tane özelliği (bağımsız değişkeni) gösteren vektör olsun. Her bir K olası sonuç yada Sk sınıfı için olasılık
$p(C_k | x_1, ... , x_n)$ ile gösterilir.

Yukarıdaki formülasyonla ilgili sorun, eğer n özelliklerinin sayısı büyükse veya bir özellik çok sayıda değer alabiliyorsa, o zaman böyle bir modeli olasılık tablolarına dayandırmak mümkün değildir. Bu nedenle, modeli daha izlenebilir hale getirmek için yeniden formüle edilebilir. Bayes Teoremini kullanarak, koşullu olasılık aşağıdaki gibi ayrıştırılabilir

```python
Image.open("Figure-1.2.jpg")
```

Uygulamada bu kesrin payı önemlidir. Çünkü payda S‟ye bağlı değildir ve sabit değerlerdir. Dolayısıyla pay kısmı $p(S_k, x_1, x_1, ... , x_n)$ bileşik olasılık modele denk gelir. Koşullu olasılık tanımı zincir kuralıyla tekrarlanırsa,

$p(S_k,x_1,...,x_n) = p(x_1,...,x_n,S_k) \\ = p(x_1 | x_2,...,x_n,S_k)p(x_2,...,x_n,S_k) \\ = p(x_1 | x_2,...,x_n,S_k)p(x_2|x_3,...,x_n,S_k)p(x_3,...,x_n,S_k) \\ = ... \\ = p(x_1 | x_2,...,x_n,S_k)p(x_2|x_3,...,x_n,S_k)...p(x_{n-1}|x_n,S_k)p(x_n|S_k)p(S_k)$

Şimdi" naive" koşullu bağımsızlık varsayımları devreye girer. X vektöründe tanımlı bütün özelliklerin Sk kategorisine koşullu olarak karşılıklı bağımsız olduğu varsayımı altında; 
$p(x_i|x_{i+1},...,x_n,S_k) = p(x_i|S_k)$ dir. Böyle bileşik model aşağıdaki gibi ifade edilir:

$p(S_k|x_1,...,x_n) \alpha p(S_k,x_1,...,x_n) \\ \quad \quad \quad \quad \quad \quad \alpha p(S_k)p(x_1|S_k)p(x_2|S_k)p(x_3|S_k)... \\ \quad \quad \quad \quad \quad \quad \alpha p(S_k) \Pi^n_{i = 1} p(x_i|S_k)$

Burada α orantıyı gösterir. S sınıf değişkeni üzerinde koşullu dağılım, $p(S_k,x_1,...,x_n) = \frac{1}{z} p(S_k) \Pi^n_{i = 1} p(x_i|S_k)$ dir. Burada $Z = p(x) = \Sigma p(S_k)p(x|S_k)$ bilinen özelliklere bağlı sabit faktör olarak isimlendirilir. Şimdiye kadar yapılan tartışma, bağımsız özellik modelini, yani naive Bayes olasılık modelini ortaya çıkarmıştır. Naive Bayes sınıflandırıcısı bu modeli bir karar kuralı ile birleştirir. Ortak bir kural, en olası hipotezi seçmektir; bu, maksimum bir sonsal veya MAP karar kuralı
olarak bilinir. Buna karşılık gelen Bayes sınırlayıcısıdır ve bazı k değerleri için aşağıdaki gibi verilen $\hat{y} = S_k$ sınıf etiketine karşılık gelen fonksiyondur.

$\hat{y} = argmaxp(S_k) \Pi^n_{i = 1} p(x_i|S_k) \\ K \in {1,...,K}$

### PYTHON’DA MULTINOMINAL NAIVE BAYES UYGULAMASI:

Daha önce açıklanan Gauss varsayımı, her bir etiket için üretken dağılımı belirtmek için kullanılabilecek tek basit varsayım değildir. Bir başka yararlı örnek, özelliklerin basit bir multinomial dağılımdan üretildiği varsayılan multinomial naive Bayes'dir. Multinomial dağılım, bir dizi kategori arasında sayımları gözlemleme olasılığını tanımlar ve bu nedenle multinomial naive Bayes, sayıları veya sayım oranlarını temsil eden özellikler için en uygun
olanıdır. Veri dağılımını en uygun Gaussian ile modellemek yerine, veri dağılımını en uygun multinomial bir dağılımla modelenir. Multinomial naive Bayes'in sıklıkla kullanıldığı önemli bir alan, özelliklerin sınıflandırılacak
belgelerdeki kelime sayımları veya frekansları ile ilişkili olduğu metin sınıflandırmasıdır. Kısa metin belgelerini, kategorilere nasıl ayrılabileceğini göstermek için 20 Newsgroups corpus'tan Kelime sayısı özelliklerini kullanacağız.

#### 20 Newsgroups veri seti

20 haber grubu veri kümesi (neredeyse) eşit olarak 20 farklı haber grupları arasında bölümlenmiş yaklaşık 20.000 haber grubu belgeleri topluluğudur. Başlangıçta Ken Lang tarafından, muhtemelen Newsweeder: Netnews kağıdını filtrelemeyi öğrenmek için toplandı, 20 haber grupları koleksiyonu, metin sınıflandırma ve metin kümeleme gibi makine öğrenme tekniklerinin metin uygulamalarında deneyler için popüler bir veri kümesi haline gelmiştir. Veriler, her biri farklı bir konuya karşılık gelen 20 farklı haber grubuna ayrılmıştır. Bazı haber grupları birbirleriyle çok yakından ilişkilidir (örneğin, comp.sys.ibm.pc.hardware/comp.sys.mac.hardware), diğerleri ise oldukça ilgisizdir (örneğin misc.forsale/soc.religion.christian). İşte konuya göre bölümlenmiş (az ya da çok) 20 haber grubunun bir listesi:

```python
Image.open("Figure-1.13.jpg")
```

Sklearn kütüphanesinde year alan; 20 haber grubu veri kümesi, iki alt kümeye bölünmüş 20 konuda yaklaşık 18.000 haber grubu yayınından oluşur: biri eğitim (veya geliştirme) ve diğeri test (veya performans değerlendirmesi) içindir. Eğitim ve test seti arasındaki bölünme, belirli bir tarihten önce ve sonra gönderilen mesajlara dayanır. sklearn.datasets.fetch_20newsgroups ifadesi verinin bulunduğu web sayfasından orijinal verileri indiren fonksiyondur. Diğer gerekli kütüphanelerle beraber veri aktarma aşağıdaki kodlarla yapılabilir.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.datasets import fetch_20newsgroups

# Veri setini yükleme
data = fetch_20newsgroups()

# Hedef kategorilerin listesini alıp yazdırma
print("Hedef Kategoriler:")
print(data.target_names)
```

Burada basitlik için, bu kategorilerden sadece beş tanesini seçeceğiz ve eğitim ve test setini indireceğiz:

```python
categories = ['misc.forsale', 'talk.religion.misc', 'soc.religion.christian', 'sci.space', 'comp.graphics']
train = fetch_20newsgroups(subset='train', categories=categories)
test = fetch_20newsgroups(subset='test', categories=categories)
```

İşte verilerden temsili bir giriş:

Örnek eğitim verisi için,

```python
print(train.data[6])
```

Örnek test verisi için,

```python
print(train.data[17])
```

Bu verileri makine öğrenimi için kullanmak için, her bir dizenin içeriğini bir sayı vektörüne dönüştürebilmemiz gerekir. Bunu yapmak için, TF-IDF vektörleştiricisini (TfidfVectorizer) kullanacağız ve onu multinomial naive Bayes sınıflandırıcısıyla ilişkilendirmek için bir pipline oluşturacağız.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

Bu pipline ile, modeli eğitim verilerine uygulayabilir ve test verileri için etiketleri tahmin edebiliriz:

```python
model.fit(train.data, train.target)
labels = model.predict(test.data)
```

Artık test verileri için etiketleri tahmin ettiğimize göre, tahmincinin performansı hakkında bilgi edinmek için bunları değerlendirebiliriz. Örneğin, test verileri için doğru ve tahmin edilen etiketler arasındaki karışıklık matrisi:

```python
from sklearn.metrics import confusion_matrix

mat=confusion_matrix(test.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, xticklabels=train.target_names, yticklabels=train.target_names)
plt.xlabel('GERÇEK ETİKET')
plt.ylabel('TAHMİN EDİLEN ETİKET')
```

Açıkçası, bu çok basit sınıflandırıcı bile uzay konuşmasını bilgisayar konuşmasından başarılı bir şekilde ayırabilir, ancak din hakkında konuşmak ve Hıristiyanlık hakkında konuşmak arasında karışır. Bu belki de beklenen bir karışıklık alanıdır. Buradaki en güzel şey, şimdi bu pipline için predict() yöntemini kullanarak herhangi bir dize için kategoriyi belirlemek için araçlara sahip olmamızdır. İşte tek bir dize için tahmini döndürecek hızlı bir yardımcı program işlevi aşağıdaki gibidir:

```python
def predict_category(s, train=train, model=model):
    tahmin = model.predict([s])
    return train.target_names[tahmin[0]]
```

```python
predict_category('I can not find CTDS')
```

```python
predict_category('discussing islam vs atheism')
```

```python
predict_category('A harmonic of the Earth\'s gravitational field ')
```
